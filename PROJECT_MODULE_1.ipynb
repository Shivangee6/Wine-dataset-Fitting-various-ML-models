<h2>Apply PCA <h2>

PCA will help us understand, the features that are mostly required and hence we can get an idea about the contribution of the number of features towards the final output.

```python
from sklearn.decomposition import PCA
from pandas import DataFrame
X = df_wine.iloc[:,0:10]
Y = df_wine.iloc[:,11]
# feature extraction
pca = PCA(n_components=4)
fit = pca.fit(X)
# summarize components
print("Explained Variance: %s" % fit.explained_variance_ratio_)
print(fit.components_)
```

```python
import matplotlib.pyplot as plt
pca = PCA(n_components=11)
pca.fit(df_wine)
pca.components_ 
columns = ['pca_%i' % i for i in range(11)]
df_pca = DataFrame(pca.transform(df_wine), columns=columns, index=df_wine.index)
df_pca.head()

import seaborn as sns
map= pd.DataFrame(pca.components_,columns=df_wine.columns)
plt.figure(figsize=(12,6))
sns.heatmap(map,cmap='twilight')
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

The above plotr shows that the majority of results can be depicted by mostly by two components alone.

```python
#We can define two components  
pca_df = PCA(n_components = 2) 
principalComponents = pca_df.fit_transform(X) 
principalDf = pd.DataFrame(data = principalComponents, columns=['principal component 1','principal component 2']) 
principalDf.head()
```

```python
#We can see the respective output associated with the components, hence can see that the quality range of 5 is seen the most.
finalDf=pd.concat([principalDf, df_wine[['quality']]],axis=1)
finalDf.head()
```

```python
fig=plt.figure(figsize=(8,8))  
ax=fig.add_subplot(1,1,1)  
ax.set_xlabel('Principal Component 1',fontsize = 15)  
ax.set_ylabel('Principal Component 2',fontsize = 15)  
ax.set_title('2 PCA Components',fontsize=20)  
targets = np.unique(df_wine['quality'])
colors=['r','g','b','m', 'y', 'k',]  
for target,color in zip(targets,colors):    
    indicesToKeep = finalDf['quality'] == target  
    ax.scatter(finalDf.loc[indicesToKeep,'principal component 1'],
              finalDf.loc[indicesToKeep,'principal component 2'],
             c=color,
             s=50)
ax.legend(targets)  
ax.grid()
```

Evaluate the performance of the model. Make sure you are testing it on the right set of data.

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    return accuracy
```
